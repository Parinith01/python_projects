```python
df = pd.read_csv('/content/sentiment_data.csv')
```

```python
df
```

```python
import nltk
nltk.download('stopwords')
stop_words = stopwords.words("english")
ps = PorterStemmer()
```

```python
def preprocessing (text):
    if not isinstance(text, str):
        return "" # Return an empty string or handle as appropriate for non-string types

    text = text.lower()

    token = nltk.word_tokenize(text)

    new = []

    for i in token:

       if i not in stop_words and i not in string.punctuation:

         stemmer = ps.stem(i)

         new.append(stemmer)

    return" ".join(new)
```

```python
import nltk
nltk.download('punkt_tab')
df["clean"] = df["Comment"].apply(preprocessing)
```

```python
df["clean"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vector = TfidfVectorizer()

X = vector.fit_transform(df["clean"])

y = df["Sentiment"]
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,

random_state=55)
```

```python
log = LogisticRegression()

log.fit(X_train,y_train)
```

```python
print("train accuracy", log.score(X_train,y_train))

print("test accuracy", log.score(X_test,y_test))
```

```python
def pred_text(text):
    clean_text = preprocessing(text)
    vectorized_text = vector.transform([clean_text])
    prediction = log.predict(vectorized_text)
    return prediction[0]
```
